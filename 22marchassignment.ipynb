```python
#Question5:
#Answer:

Age = [25,42,37,19,31,28]
softdrink = ["coke","pepsi","mountain dew","coke","pepsi","coke"]
```

```python
data = pd.DataFrame({"Age(Years)": Age ,"Soft drink Preference":softdrink} )
```

```python
data
```

```python
preference_mapping = {'coke': 0, 'pepsi': 1, 'mountain dew': 2}
```

```python
data["Soft drink Preference"] = data["Soft drink Preference"].map(preference_mapping)
```

```python
data
```

```python
data.corr(method="pearson")
```
